# **Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.pie( df['Uni Student'].value_counts(), labels=df['Uni Student'].value_counts().index,autopct='%1.2f%%')

plt.title("University Students")
plt.show()
```

```python
df['Age'].value_counts().plot(kind='bar')
plt.title("Age ")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
df['Sleep_per_night'].value_counts().plot(kind='bar')
plt.title("Sleep Hours per Night")
plt.xlabel("Hours")
plt.ylabel("Count")
plt.show()
```

```python
df['Sleep_quality'].value_counts().plot(kind='bar')
plt.title("Sleep Quality")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
df['Caffeine_per_day'].value_counts().plot(kind='bar')
plt.title("Caffeine Consumption per Day")
plt.xlabel("Drinks per Day")
plt.ylabel("Count")
plt.show()
```

```python
df['Caffeine_time'].value_counts().plot(kind='bar')
plt.title("Time of Caffeine Consumption")
plt.xlabel("Time")
plt.ylabel("Count")
plt.show()
```

```python
plt.pie(df['Caffeine_4hr_before_sleep'].value_counts(),labels=df['Caffeine_4hr_before_sleep'].value_counts().index,  autopct='%1.2f%%')

plt.title("Caffeine Before Sleep")
plt.show()
```

```python
plt.pie( df['Caffeine_affects_sleep'].value_counts(),labels=df['Caffeine_affects_sleep'].value_counts().index,autopct='%1.2f%%')

plt.title("Does Caffeine Affect Sleep?")
plt.show()
```
